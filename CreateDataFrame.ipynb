```python
import geocoder
from ipynb.fs.full.ScrapePostCode import df #import df from ScrapePostCode file
```

```python
df
```

```python
postal_code = df.loc[:,'Postcode']
```

```python
postal_code
```

```python
length = len(df)
length
```

```python
lat_lng_coords = [None] * length # Initialize the list lat_lng_coords of size length to the list of None
latitude = []
longitude = []
```

```python
i=0
while(lat_lng_coords[i] is None) and (i<length):
    g = geocoder.arcgis('{}, Toronto, Ontario'.format(postal_code[i]))
    lat_lng_coords[i] = tuple(g.latlng)
    latitude.append(lat_lng_coords[i][0])
    longitude.append(lat_lng_coords[i][1])
    i = i + 1
    if i == length:
        break

df['Latitude'] = latitude
df['Longitude'] = longitude
```
